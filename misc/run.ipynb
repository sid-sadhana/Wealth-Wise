```python

# %%
from pypdf import PdfReader
from groq import Groq

# PDF Extraction with Improved Handling
reader = PdfReader("test.pdf")
number_of_pages = len(reader.pages)

# Efficient text accumulation with list join (better than `+=`)
text_list = []
for i in range(number_of_pages):
    page = reader.pages[i]
    text = page.extract_text() or ""  # Handle cases where text extraction fails
    text_list.append(text.strip())

out = " ".join(text_list)
import re

def remove_non_english_words(text):
    # Regex pattern: Matches words with English letters or numbers
    return ' '.join(re.findall(r'\b[a-zA-Z0-9]+\b', text))


# Example Usage
cleaned_text = remove_non_english_words(out)
```
